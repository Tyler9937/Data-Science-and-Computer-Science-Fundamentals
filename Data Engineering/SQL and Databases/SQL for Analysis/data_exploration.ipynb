```python
import pandas as pd

data_path = 'data/mr_beast_youtube_stats.csv'
df = pd.read_csv(data_path)

for index, row in df.iterrows():
    title = row.title.replace("'", "")
    title = title.replace('"', '')
    description = row.description.replace("'", "")
    description = description.replace('"', '')
    print(title)
    print(description)
```
